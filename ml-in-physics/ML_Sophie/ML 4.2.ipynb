### XOR in Keras
- One hidden layer is enough to realize this network: an input layer, one hidden layer and an output layer.
- using supervised learning approach to solve X-OR using neural network


```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
# XOR: 4 examples, 2 features

# input and expected output
x = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,1,1,0])

# The activation function in output layer is selected based on the output space. 
# For a binary classification task sigmoid activations is correct choice 
# for multi class classification softmax is the most populary choice.

model = Sequential()

# hidden unit
model.add(Dense(units=2,activation="relu", input_dim=2))

# output layer 
model.add(Dense(units=1,activation="sigmoid"))

# compile model
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])

# minimize the cost function. This is achieved using back propagation algorithm.
model.fit(x,y,epochs=1000,batch_size=4)



# Notes: 
# We can get weight value in keras using model.get_weights() function.
# Selection of a loss and cost functions depends on the kind of output we are targeting. 
# For classification we use cross entropy cost function. 
# In Keras we have binary cross entropy cost funtion for binary classification
```

```python
# What are the network parameters after training?

print(model.layers[0].get_weights()) # hidden layer
print(model.layers[1].get_weights()) # output layer 
```
