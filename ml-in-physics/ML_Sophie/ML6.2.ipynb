# AlexNet
- solves the problem of image classification where the input is an image of one of 1000 different classes (e.g. cats, dogs etc.) and the output is a vector of 1000 numbers
- ith element of the output vector is interpreted as the probability that the input image belongs to the ith class- sum of all elements of the output vector is 1

- convolutional layers which are followed by 3 fully connected layers.
- with Relu applied after each of them, and dropout applied before the first and second fully connected layer.

```python
# from 5.1 

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Input, Dropout, Activation
from tensorflow.keras.layers import Conv2DTranspose, MaxPooling2D, Conv2D, Conv3D
from tensorflow.keras.layers import concatenate
# imports our layers we want to use

from tensorflow.python.keras.losses import categorical_crossentropy #loss function
from tensorflow.python.keras.optimizers import Adam, SGD #optimisers
from tensorflow.keras.utils import to_categorical #some function for data preparation

from tensorflow.compat.v1 import disable_eager_execution
from tensorflow.python.keras.optimizers import RMSprop
```

```python
# build model

# Max-pooling layers, 
# of the kind described in Section 3.4
# follow both response-normalization layers 
# as well as the fifth convolutional layer
# This is what we use throughout our network, with s = 2 and z = 3.

# a pooling layer can be thought of as 
# consisting of a grid of pooling units spaced s pixels apart, 
# each summarizing a neighborhood of size z × z 
# centered at the location of the pooling unit.

# ReLU
# ReLU non-linearity is applied to the output of every convolutional
# and fully-connected layer.

model = Sequential() # empty model

# first convolutional layer 
# filters the 224×224×3 input image with 96 kernels of size 11×11×3 with a stride of 4 pixels
model.add(Conv2D(filters=48, kernel_size=(11,11), input_shape=(224, 224, 3), strides =(4,4), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(3,3), strides=(2,2), padding='same'))

# second convolutional layer 
# takes as input the (response-normalized and pooled) output of the first convolutional layer
# filters it with 256 kernels of size 5 × 5 × 48.
model.add(Conv2D(256, (5,5), strides =(1,1), padding='valid', activation='relu'))
model.add(MaxPooling2D(pool_size=(3,3), strides=(2,2), padding='valid'))

# third, fourth, and fifth convolutional layers 
# connected to one another without any interveng inpooling or normalization layers

# third convolutional layer 
# has 384 kernels of size 3 × 3 × 256 connected to the (normalized, pooled) outputs of the second convolutional layer
model.add(Conv2D(384, (3,3), strides =(1,1), padding='valid', activation='relu'))

# fourth convolutional layer 
# has 384 kernels of size 3 × 3 × 192
model.add(Conv2D(384, (3,3), strides =(1,1), padding='valid', activation='relu'))

# fifth convolutional layer 
# has 256 kernels of size 3 × 3 × 192
model.add(Conv2D(256, (3,3), strides =(1,1), padding='valid', activation='relu'))
model.add(MaxPooling2D(pool_size=(3,3), strides=(2,2), padding='valid'))

# # The fully-connected layers have 4096 neurons each.
# first dense layer
model.add(Flatten())
model.add(Dense(4096,input_shape=(224*224*3,), activation='relu'))
model.add(Dropout(0.5))

# second dense layer
model.add(Dense(4096, activation='relu'))
model.add(Dropout(0.5))

# output
model.add(Dense(10000, activation='softmax'))
        
model.compile(loss=categorical_crossentropy, optimizer="adam", metrics=["accuracy"])
model.summary()   
    
```
