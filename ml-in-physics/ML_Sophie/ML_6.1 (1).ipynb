```python
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf

from tensorflow.keras.datasets import mnist #to import our dataset
from tensorflow.keras.models import Sequential, Model # imports our type of network
from tensorflow.keras.layers import Dense, Flatten,Conv2D, Input, Lambda, Layer, ReLU, MaxPooling2D, Dropout, Conv2DTranspose, concatenate, Input # imports our layers we want to use

from tensorflow.python.keras.losses import categorical_crossentropy #loss function
from tensorflow.python.keras.optimizers import Adam, SGD #optimisers
from tensorflow.python.keras.utils.np_utils import to_categorical #some function for data preparation
```

```python
filters = 64

i = Input(shape=(256,256,1))
x = Conv2D(filters,(3,3),padding="same",activation="relu")(i)
x = Conv2D(filters,(3,3),padding="same",activation="relu")(x)
max1 = MaxPooling2D((2,2))(x)
drop1 = Dropout(0.5)(max1)

x2 = Conv2D(filters*2,(3,3),padding="same",activation="relu")(drop1)
x2 = Conv2D(filters*2,(3,3),padding="same",activation="relu")(x2)
max2 = MaxPooling2D((2,2))(x2)
drop2 = Dropout(0.5)(max2)

x3 = Conv2D(filters*4,(3,3),padding="same",activation="relu")(drop2)
x3 = Conv2D(filters*4,(3,3),padding="same",activation="relu")(x3)
max3 = MaxPooling2D((2,2))(x3) 
drop3 = Dropout(0.5)(max3)

x4 = Conv2D(filters*8,(3,3),padding="same",activation="relu")(drop3)
x4 = Conv2D(filters*8,(3,3),padding="same",activation="relu")(x4)
max4 = MaxPooling2D((2,2))(x4)
drop4 = Dropout(0.5)(max4)

##############

x5 = Conv2D(filters*16,(3,3),padding="same",activation="relu")(drop4)
x5 = Conv2D(filters*16,(3,3),padding="same",activation="relu")(x5)

y1 = Conv2DTranspose(filters*8,(3,3),strides=(2,2),padding="same",activation="relu")(x5)
c1 = concatenate([y1,x4])
drop5 = Dropout(0.5)(c1)
x6 = Conv2D(filters*8,(3,3),padding="same",activation="relu")(drop5)
x6 = Conv2D(filters*8,(3,3),padding="same",activation="relu")(x6)

y2 = Conv2DTranspose(filters*4,(3,3),strides=(2,2),padding="same",activation="relu")(x6)
c2 = concatenate([y2,x3])
drop6 = Dropout(0.5)(c2)
x7 = Conv2D(filters*4,(3,3),padding="same",activation="relu")(drop6)
x7 = Conv2D(filters*4,(3,3),padding="same",activation="relu")(x7)

y3 = Conv2DTranspose(filters*2,(3,3),strides=(2,2),padding="same",activation="relu")(x7)
c3 = concatenate([y3,x2])
drop7 = Dropout(0.5)(c3)
x8 = Conv2D(filters*2,(3,3),padding="same",activation="relu")(drop7)
x8 = Conv2D(filters*2,(3,3),padding="same",activation="relu")(x8)

y4 = Conv2DTranspose(filters*2,(3,3),strides=(2,2),padding="same",activation="relu")(x8)
c4 = concatenate([y4,x])
drop8 = Dropout(0.5)(c4)
x9 = Conv2D(filters*2,(3,3),padding="same",activation="relu")(drop8)
x9 = Conv2D(filters*2,(3,3),padding="same",activation="relu")(x9)

x10 = Conv2D(1, (1,1), padding="same", activation="sigmoid")(x9)


model = Model(i,x10,name="unet")
```
