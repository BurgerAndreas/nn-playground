# XOR in Keras
A single layer Perceptron cannot accommodate the XOR function. Show that by utilising a single hidden dense layer with
sigmoid activation functions, XOR can be realised. Central takeaways from this exercise are
* How to implement this network utilising tensorflow.keras.
* How to analyse the trained network?
* What are the theoretical constraints to the size of the network and what are practical sizes needed for good performance?

```python
import numpy as np
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
# Define input and output values according to the XOR truth table. Nothing for you to do here.
xor_input = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], "float32")
xor_output = np.array([[0], [1], [1], [0]], "float32")
```

Add the hidden dense layer, and the output layer to the `xor_model` object with sigmoid activation function. Use `16`
units in the hidden layer.

Then compile the model using binary crossentropy as loss, Adam as optimizer, and accuracy as metric.

Finally train the model for `2000` epochs with batch size `4`. Store the training history in history.

```python
history = None
xor_model = Sequential()

hidden_units = 16
activation = 'sigmoid'

#Hidden Layer
xor_model.add(Dense(units=hidden_units, activation=activation))

#Output Layer
xor_model.add(Dense(units=1, activation=activation))

optimizer = 'adam'
loss = 'binary_crossentropy'
metrics = ['accuracy']

#Compile model
xor_model.compile(optimizer=optimizer, loss=loss, metrics=metrics)

epochs = 2000
batch_size = 4

#Train model 
history = xor_model.fit(xor_input, xor_output, batch_size=batch_size, epochs=epochs)
```

Print out the model summary, the model weights, the final accuracy and the prediction for all input values.

```python
print(xor_model.summary(),'\n')
print(xor_model.get_weights(),'\n')
print(history.history['accuracy'][-1],'\n')
print(xor_model.predict(xor_input),'\n')
```
