# Classifiers in sklearn 
The aim of this exercise is to familiarise yourself with some of the classifiers implemented in sklearn
by applying them to a 2D classification task.<br>
The class Dataset is used to generate point-clouds with labels $\pm$1.
* Using such datasets, build classifiers with svm.svc, DecisionTreeClassifier, RandomForestClassifier, AdaBoostClassifier.
* Using default parameters (except max_depth), check their performance on your pointcloud by checking their score
and visualising their classification boundaries.
* Think about how your results for the DecisionTreeClassifier change when varying the maxdepth.

````python

import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.exceptions import NotFittedError
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
````

````python
class Dataset:
    """
    Class creating the dataset. Nothing for you to do here.
    """

    def __init__(self, length: int = 1000):
        self.length = length
        self.xyset = 3. * (np.random.rand(self.length, 2) - [0.5, 0.5])
        self.classes = [self.con1(self.xyset[i, 0], self.xyset[i, 1]) +
                        self.con2(self.xyset[i, 0], self.xyset[i, 1]) +
                        self.con3(self.xyset[i, 0], self.xyset[i, 1]) for i in range(self.length)]
        self.x_train = None
        self.y_train = None
        self.x_test = None
        self.y_test = None

    def con1(self, x, y):
        if y >= 0.2 * np.sin(10 * x) + x:
            rv = 1
        else:
            rv = -1
        return rv

    def con2(self, x, y):
        if (x + 0.5) ** 2 + (y - 0.5) ** 2 <= 0.1:
            rv = -2
        else:
            rv = 0
        return rv

    def con3(self, x, y):
        if (x - 0.5) ** 2 + (y + 0.5) ** 2 <= 0.1:
            rv = 2
        else:
            rv = 0
        return rv

    def plotTrueBoundaries(self, ax):
        phiRange = np.arange(0, 7, 0.1)
        xRange = np.arange(-1.5, 1.5, 0.01)
        x1 = -0.5 + np.sqrt(0.1) * np.cos(phiRange)
        y1 = 0.5 + np.sqrt(0.1) * np.sin(phiRange)
        x2 = 0.5 + np.sqrt(0.1) * np.cos(phiRange)
        y2 = -0.5 + np.sqrt(0.1) * np.sin(phiRange)
        y3 = 0.2 * np.sin(10 * xRange) + xRange
        ax.plot(x1, y1, 'black')
        ax.plot(x2, y2, 'black')
        ax.plot(xRange, y3, 'black')

    def plot_dataset_with_boundaries(self):
        fig = plt.figure(figsize=(7, 7))
        fig.patch.set_facecolor("white")
        plt.ticklabel_format(useOffset=False)

        plt.scatter(self.xyset[:, 0], self.xyset[:, 1], c=self.classes)
        self.plotTrueBoundaries(plt.gca())
        plt.gca().set_aspect('equal')
        return plt.gca()

    def train_test_split(self, test_fraction=0.25):
        self.x_train, self.x_test, self.y_train, self.y_test = train_test_split(self.xyset, self.classes,
                                                                                test_size=test_fraction)



````

````python
class Classifier:

    def __init__(self, classifier_type, dataset, max_depth=None):
        """
        :param classifier_type: determine which type of sklearn classifier to use
        :type classifier_type: str with one of the values ["DecisionTree", "SVC", "RandomForest", "AdaBoost"]
        :param dataset: the dataset to fit on
        :type dataset: Dataset
        :param max_depth: maximum depth of classifier (only applicable to some of the classifiers)
        :type max_depth: int or None
        """
        self.classifier_type = classifier_type
        self.dataset = dataset
        self.y_test = dataset.y_test
        self.x_test = dataset.x_test
        self.y_train = dataset.y_train
        self.x_train = dataset.x_train
        self.max_depth = max_depth
        self.train_score = None
        self.test_score = None

        self.build_model()

    def build_model(self):
        """
        Depending on self.classifier_type (classifier type can take the values "DecisionTree", "SVC", "RandomForest", "AdaBoost")
        initialize the correct model from sklearn (SVC Classifier, DecisionTree Classifier, RandomForest Classifier and
        AdaBoost Classifier: see imports) and pass the parameter self.max_depth to the appropriate input of the
        Classifier, where possible. Write the resulting model instance to self.clf. In case  This function should not
        return anything. If self.classifier_type does not match any of the known classifiers, raise a ValueError with 
        a sensible error message.
        """
        if self.classifier_type == 'DecisionTree':
            self.clf = DecisionTreeClassifier(max_depth=self.max_depth)
        elif self.classifier_type == 'SVC':
            self.clf = SVC()
        elif self.classifier_type == 'RandomForest':
            self.clf = RandomForestClassifier(max_depth=self.max_depth)
        elif self.classifier_type == 'AdaBoost':    
            self.clf = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=self.max_depth))
        else:
            raise ValueError('ClassifierType does not exist or is not supported.')

    def fit(self):
        '''
        Fit the classifier to the training data set and store the training and test score of the model in
        self.train_score and self.test_score.
        '''
        self.clf.fit(self.x_train, self.y_train)
        self.train_score = self.clf.score(self.x_train, self.y_train)
        self.test_score = self.clf.score(self.x_test, self.y_test)

    def plot_classifier_boundary(self, axis):
        '''
        plots the decision boundary of the classifier. Nothing for you to do here.
        '''
        # Get boundaries of training data
        x_1_train_min = min(self.x_train[:, 0])
        x_2_train_min = min(self.x_train[:, 1])
        x_1_train_max = max(self.x_train[:, 0])
        x_2_train_max = max(self.x_train[:, 1])

        x_1_grid, x_2_grid = np.meshgrid(np.arange(x_1_train_min, x_1_train_max, 0.1),
                                         np.arange(x_2_train_min, x_2_train_max, 0.1))

        grid_points = np.c_[x_1_grid.ravel(), x_2_grid.ravel()]

        predictions = self.clf.predict(grid_points).reshape(len(x_1_grid), -1)

        # contour plot to show the boundary
        axis.contour(x_1_grid, x_2_grid, predictions, levels=[1, -1], colors='g', linestyles='solid', linewidths=3)

    def plot_dataset_with_clf_boundary(self):
        '''
        plots the dataset with the actual boundaries and the decision boundary found by the classifier. Nothing for you
        to do here
        '''
        ax = self.dataset.plot_dataset_with_boundaries()
        self.plot_classifier_boundary(ax)
        if self.max_depth is not None:
            plt.title(self.classifier_type + " with max_depth " + str(self.max_depth))
        else:
            plt.title(self.classifier_type)
````

````python
training_scores = np.array([])
test_scores = np.array([])

max_depth = 3
typeArray = ['DecisionTree', 'SVC', 'RandomForest', 'AdaBoost']

for classifier_type in typeArray:
    classifier = Classifier(classifier_type=classifier_type, dataset=dataset, max_depth=max_depth)
    classifier.fit()
    classifier.plot_dataset_with_clf_boundary()
    training_scores = np.append(training_scores, classifier.train_score)
    test_scores = np.append(test_scores, classifier.test_score)
    
#print(training_scores)
#print(test_scores)
````
