```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
from sklearn.kernel_ridge import KernelRidge
#from sklearn.metrics.pairwise import polynomial_kernel
```


- Consider the underlying model $y = 2x$. Generate data with some Gaussian noise (100 datapoints). Solve the optimisation problem for $\lambda = 0, 0.1, 1$ as described in the lectures.

Since <br>
$$C(w) = \frac{1}{2}\sum_{n=1}^{100}(w x_n-y_n)^2+\frac{\lambda}{2}w^2$$
with $y_n = 2x_n $ <br> one can calculate the minimum by setting
$$0 = \frac{\partial C}{\partial w}=\sum_{n=1}^{100}(w x_n-y_n)x_m+\lambda w = \vec{x}^T \vec{x}w+\lambda w - \vec{x}^T\vec{y}.$$
<br> The solution is <br>
$$w_R = \frac{\vec{x}^T \vec{y}}{\vec{x}^T \vec{x}+\lambda}=\frac{x_1y_1+x_2y_2+...+x_{100}y_{100}}{x_1x_1+x_2+...+x_{100}x_{100}+\lambda}$$

```python
def gen_data(a, b, num_points):
    '''The underlying model is y = 2x. Data with some Gaussian noise is generated for num_points datapoints in the interval
    a to b.'''
    x = np.linspace(a, b, num_points)
    noise = np.random.normal(loc = 0, scale = 0.5, size = num_points)
    y = 2 * x + noise
    return x, y

# calculate weights = coeff
def calculate_w(lamb, x, y):
    xTx = x.transpose().dot(x)
    m = xTx + lamb
    xTy = x.transpose().dot(y)
    w = xTy/m
    return w
    

# Generate the data
x, y = gen_data(1, 100, 100)


# Optimize for lambda = 0, 0.1, 1. and plot it
lambdas = [0, 0.1, 1]
w = []
for l in lambdas:
    w.append(calculate_w(l, x, y))
    label = "lambda = "+str(l)
    plt.plot(x, calculate_w(l, x, y)*x, label = label)

plt.scatter(x,y, c = "red")
plt.legend()
plt.title(" ")
plt.show()
print("w for lambda = 0, 0.1, 1:")
print(w)

# compare the results for w to the sklearn implementation
w_sklearn = []
for l in lambdas:
    model = Ridge(alpha = l, fit_intercept=False)
    f = model.fit(x[:, np.newaxis], y) # is f used?
    w_sklearn.append(model.coef_[0])

print("compare to sklearn implementation: w for lambda = 0, 0.1, 1:")
print(w_sklearn)
```

- We now introduce a kernel $\phi(x) = (x, x^2)$.
Solve again the problem for the underlying model $y = 2x$.
Note that this time you need to recalculate the target values $t_n$ appropriately.

Now
$$C(\mathbf{w}) = \frac{1}{2}\sum_{n=1}^{100}(\mathbf{w}^T \phi(x_n)-t_n)^2+\frac{\lambda}{2}\mathbf{w}^T\mathbf{w}$$
<br> and <br>
$$
\mathbf{w_R} = (\mathbf{\phi}^T\mathbf{\phi}+\lambda \mathbf{1}_{2x2})^{-1}\mathbf{\phi}^T\mathbf{t} =
\left[ \left( \begin{matrix}  x_1^2+x_2^2+...+x_{100}^2 & x_1^3+x_2^3+...+x_{100}^3 \\
 x_1^3+x_2^3+...+x_{100}^3 & x_1^4+x_2^4+...+x_{100}^4  \end{matrix} \right)+\lambda \mathbf{1}_{2x2}\right]^{-1}
\left( \begin{matrix}  x_1(y_1+y_1^2)+x_2(y_2+y_2^2)+...+x_{100}(y_{100}+y_{100}^2)  \\
 x_1^2(y_1+y_1^2)+x_2^2(y_2+y_2^2)+...+x_{100}^2(y_{100}+y_{100}^2)  \end{matrix} \right)
 $$
<br>
where <br>
$\mathbf{\phi}= ((x_1, x_1^2), ..., (x_{100}, x_{100}^2))^T$ and $\mathbf{t}= (y_1, ..., y_{100})^T$?

```python
def recalculate(x, y):
    '''Calculate phi and the new target values.'''
    phi = []
    t = []

    for i in range(len(x)):
        phi.append([x[i], x[i]**2])
        #t.append([y[i] + y[i]**2])
        t.append([y[i]])

    phi = np.array(phi)
    t = np.array(t)
    return phi, t


def calculate_w_2(lamb, x, y):
    '''Calculate w, phi and the new target values.'''
    phi, t = recalculate(x,y)

    phiT = phi.transpose()
    phiTphi = phiT.dot(phi)
    phiTt = phiT.dot(t)

    m = phiTphi + np.array([[lamb, 0], [0, lamb]])
    m_inv = np.linalg.inv(m)

    w = m_inv.dot(phiTt)
    return phi, t, w

# calculate w2 for kernel problem
w2 = []
for l in lambdas:
    phi, t, w = calculate_w_2(l, x, y)
    w2.append(w)

print("w for lambda = 0, 0.1, 1:")
print(w2)


#compare the results to the sklearn implementation
def calculate_kernel_ij(x1, x2): 
    """function which is used in KernelRidge of sklearn"""
    phi_i = np.array([x1, x1**2])
    phi_iT = phi_i.transpose()
    phi_j = np.array([x2, x2**2])
    k_ij = phi_iT.dot(phi_j)
    return k_ij

w2_sklearn = []
pred_sklearn = []
pred = []
j = 0

for l in lambdas:
    # create model KernelRidge and fit it
    model = KernelRidge(alpha = l, kernel = calculate_kernel_ij, degree = 2)
    x1 = x[:,np.newaxis]
    y1 = y[:, np.newaxis]
    f = model.fit(x1, y1) # not used?
    w2_sklearn.append(model.dual_coef_) # the weights calculated by sklearn in kernel space (thats why they cant be compared directly to my values)

    # calculate the predictions for the sklearn weights and my weights
    pred_sklearn.append(model.predict(x1)) 
    pred.append(w2[j].transpose().dot(phi.transpose())) # phi defined in calculate_w_2()
    j += 1

    # plot the predictions for my w
    label = "lambda = "+str(l)
    plt.plot(x1, pred[0].transpose(), label = label)

plt.scatter(x,y, c = "red")
plt.title(" ")
plt.legend()
plt.show()

#compare the results by calculating normpredictions from sklearn - my predictions)
print("compare to sklearn implementation: norm(predictions from sklearn - my predictions) for lambda = 0, 0.1, 1:")
print([np.linalg.norm(pred[j]-pred_sklearn[j].transpose()) for j in range(3)])


#print(phi)
```

- Calculate the kernel function $k(x_n, x_m) = \phi(x_n)^T \phi(x_m)$.

```python
kernel = (phi).dot(np.transpose(phi))
print(kernel)
```

- Using this kernel function solve the optimisation problem for the dual variables $\mathbf{a}$.

With $\mathbf{w} = \phi^T\mathbf{a}$ one gets <br>
$$C(\mathbf{a})=\frac{1}{2}\mathbf{a}^T \mathbf{K}\mathbf{K}\mathbf{a}-\mathbf{a}^T\mathbf{K}\mathbf{t}+\frac{1}{2}\mathbf{t}^T\mathbf{t}+\frac{\lambda}{2}\mathbf{a}^T\mathbf{K}\mathbf{a},$$
<br>
which is minimized for <br>
$$\mathbf{a}= \left( \mathbf{K}+\lambda \mathbf{1}_{100x100}\right)^{-1}\mathbf{t} $$


```python
a = []
lambdas = [0, 0.1, 1]

# calculate a
for l in lambdas:
    m = kernel + np.identity(len(x))*l
    m_inv = np.linalg.inv(m)
    a_l = m_inv.dot(t) # t defined in recalculate()
    a.append(a_l)
    print("for lambda = "+str(l)+":")
    print(a_l)
```

- Show that both solutions agree by obtaining $\mathbf{w} = \phi^T\mathbf{a}$ from your solution $\mathbf{a}$.

```python
w_dual = []

# weights from a
for a_l in a:
    w_froma = phi.transpose().dot(a_l)
    w_dual.append(w_froma)
    
print("compare results:")
print("from Point 2:")
print(w2)
print("from a:")
print(w_dual)
```

The results for $\lambda = 0.1$ and $\lambda = 1$ agree. For $\lambda = 0$ gives not the right results.
