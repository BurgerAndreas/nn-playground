# Kernel methods (peer correction)
The aim of this exercise is to familiarise yourself with the dual optimisation problem.


* We obtain the solution to a linear regression with respect to our weights $w$ as previously discussed in the lectures.
$$C(w)=\frac{1}{2}\sum_{n=1}^N \left({\bf w}^Tx_n-y_n\right)^2+\frac{\lambda}{2}{\bf w}^T{\bf w}$$
To implement this, consider the underlying model $y=2x.$ Generate data with some Gaussian noise (100 datapoints). Solve the optimisation problem for $\lambda=0,0.1,1$ as described in the lectures.
* We now introduce a kernel $\phi(x)=(x,x^2).$ Solve again the problem for the underlying model $y=2x.$ Note that this time you need to recalculate the target values $t_n$ appropriately.
* Calculate the kernel function $k(x_n,x_m)=\phi(x_n)^T \phi(x_m).$
* Using this kernel function solve the optimisation problem for the dual variables ${\bf a}.$
* Show that both solutions agree by obtaining ${\bf w}$ from your solution ${\bf a}.$


```python
from typing import Union

import matplotlib.pyplot as plt
import numpy as np
```

```python
def generate_data(num_samples=100):
    x = np.linspace(-1, 1, num_samples)
    # To illustrate the equivalence we set the variance of the noise rather low. It works just
    # as well for nosier data, but 100 samples are a bit too few to see it well.
    y = 2 * x + np.random.normal(size=num_samples, scale=0.1)
    return x, y
```

Since the cost function given above is the one for Ridge Regression, we know the solution to be
$\hat{w}=\left(X^{T}X+\lambda\boldsymbol{1}\right)^{-1}X^{T}y$

In our one dimensional the formula rather simplifies and is implemented below. We did not
introduce an offset here, but did below.


```python
class RidgeRegressor:
    def __init__(self, X, y, type: Union["simple", "dual", "kernel"] = "simple"):
        self.X = X
        self.y = y

        self.type = type

        self.weights = None
        self.weights_record = []

        self.a = None
        self.a_record = []

        self.phi = None
        self.t = None
        self.kernel = None

    def transform_to_higher_dimension(self):
        self._transform_to_feature_vector()
        self._calculate_kernel()

    def fit_data(self, regulator=0.1):
        if self.type == "simple":
            self._fit_1d_data(regulator)
        elif self.type == "dual":
            self._fit_phi(regulator)
        elif self.type == "kernel":
            self._fit_kernel_method(regulator)

    def plot_fitted_model(self):
        if self.type == "simple":
            fig = plt.figure(figsize=(10, 10))
            fig.patch.set_facecolor("white")
            plt.scatter(self.X, self.y)
            plt.plot(self.X, np.dot(self.X, self.weights))
            plt.xlabel("X")
            plt.ylabel("y")
            plt.title("Data and Fit from Simple Ridge Regression")
            plt.show()

    def _transform_to_feature_vector(self):
        self.phi = np.vstack([self.X, self.X ** 2]).transpose()
        self.t = self.y + self.y ** 2

    def _calculate_kernel(self):
        self.kernel = np.dot(self.phi, self.phi.transpose())

    def _fit_1d_data(self, regulator):
        data_matrix = np.dot(self.X.transpose(), self.X)
        self.weights = 1 / (data_matrix + regulator) * np.dot(self.X.transpose(), y)
        self.weights_record.append(self.weights)

    def _fit_phi(self, regulator):
        data_matrix = np.dot(self.phi.transpose(), self.phi)
        inverse = np.linalg.inv(data_matrix + regulator * np.eye(data_matrix.shape[0]))
        self.weights = np.dot(inverse, np.dot(self.phi.transpose(), self.t))
        self.weights_record.append(self.weights)

    def _fit_kernel_method(self, regulator=0.1):
        inverse = np.linalg.pinv(self.kernel + np.identity(self.kernel.shape[0]) * regulator)
        self.a = np.dot(inverse, self.t)
        self.a_record.append(self.a)
        self._transform_weights()

    def _transform_weights(self):
        self.weights = np.dot(self.phi.transpose(), self.a)
        self.weights_record = [np.dot(self.phi.transpose(), a) for a in self.a_record]
```

Below we generate the noisy data for the model $y=2x$


```python
x, y = generate_data()
```

First we apply Ridge Regression in one dimension for $y=2x$.

```python
one_dimensional_regressor = RidgeRegressor(x, y, type="simple")
for regulator in [0, 0.1, 1]:
    one_dimensional_regressor.fit_data(regulator)
    one_dimensional_regressor.plot_fitted_model()

print("The weights are: ", one_dimensional_regressor.weights_record)
```

Now we use $\phi\left(x\right)=\left(x,x^{2}\right)$ and the target variable $t=\left(y,y^{2}\right)$.
Applying Ridge Regression gives the following weights.

```python
two_dimensional_regressor = RidgeRegressor(x, y, type="dual")
for regulator in [0, 0.1, 1]:
    two_dimensional_regressor.transform_to_higher_dimension()
    two_dimensional_regressor.fit_data(regulator)

print("The weights are: \n", two_dimensional_regressor.weights_record)
```

```python
kernel_regressor = RidgeRegressor(x, y, type="kernel")
for regulator in [0, 0.1, 1]:
    kernel_regressor.transform_to_higher_dimension()
    kernel_regressor.fit_data(regulator)

print("The weights are: \n", kernel_regressor.weights_record)
```

```python
assert np.all([np.isclose(two_dimensional_weights - kernel_weights, 0) for two_dimensional_weights, kernel_weights in
               zip(two_dimensional_regressor.weights_record,
                   kernel_regressor.weights_record)]), "The weights do not agree"
print("The weights agree")
```
